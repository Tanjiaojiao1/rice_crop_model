```python
groups = df.groupby('DStage')

# 设置标记列表以便用于不同的 fun 值
markers = ['o', 's', '*', '^', 'v', '>', '<', 'D', 'P', 'X', '1','_']

# 创建一个大图和 2x3 个子图对象
fig, axs = plt.subplots(2, 3, figsize=(15, 8))

# 将 axs 数组展平，以便在遍历子图时使用单个索引
axs = axs.flatten()

# 定义各个 DStage 的顺序
dstage_order = ['tillering date', 'jointing date', 'booting date', 'heading date', 'maturity date']

# 遍历每个 DStage 分组，按照指定的顺序绘制子图
for i, dstage in enumerate(dstage_order):
    # 获取当前 DStage 分组的数据
    group = groups.get_group(dstage)
    # 获取当前分组的 fun 值
    fun_values = group['fun'].unique()
    # 遍历每个 fun 值
    for j, fun_value in enumerate(fun_values):
        # 获取当前 fun 值的数据
        sub_group = group[group['fun'] == fun_value]
        # 获取当前 fun 值对应的颜色和标记
        marker = markers[j % len(markers)]
        if j <9:
            # 绘制散点图
            axs[i].scatter(sub_group['Date_DOY'], sub_group['Date_sim_DOY'], marker=marker, color='#FF6666', label=fun_value,s=10)
            axs[i].scatter(sub_group['Date_DOY'], sub_group['Date_simphoto_DOY'], marker=marker, color= '#3399FF', label=fun_value,s=6)
        else:
            axs[i].scatter(sub_group['Date_DOY'], sub_group['Date_sim_DOY'], marker=marker, color='#FF6666', label=fun_value,s=6, alpha=0.5)
#         # 在第一行第一列的子图中添加图例
#             if i == 5 and j == 11:
#                 axs[i].legend(ncol=4, bbox_to_anchor=(0, -0.2), loc='upper left', frameon=False)
    # 设置横纵坐标轴标签
    axs[i].set_xlabel('Date_DOY')
    axs[i].set_ylabel('Date_sim_DOY or Date_simphoto_DOY')
    # 设置子图标题
    axs[i].set_title(dstage)

# 删除第二行第三列的子图
fig.delaxes(axs[-1])

# 在第二行第三列的子图位置添加第一行第一列子图的图例
axs[-2].legend(*axs[0].get_legend_handles_labels(), ncol=4, bbox_to_anchor=(0, -0.2), loc='upper left', frameon=False)

# 调整子图之间的间距
plt.subplots_adjust(wspace=0.3, hspace=0.4)

# 显示图形
plt.show()
```
