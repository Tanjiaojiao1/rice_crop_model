```python
dfall = pd.DataFrame()
dir_path = 'D:/workspace/rice_crop_model/data/clusters/sites_clusters/general_parameters/result_3'
for root, dirs, files in os.walk(dir_path):
    for file in files:
        file_path = os.path.join(root, file)
        funname = file_path[92:-6]
        data = pd.read_excel(file_path)
        data = data.loc[data.DStage != 'reviving date']
#         data = data.loc[data.DStage != 'maturity date']
        data.reset_index(inplace=True)
        data = data.iloc[:, 7:]
        data.drop(columns=['Date', 'Date_simphoto', 'Date_sim'], axis=1, inplace=True)
        data = pd.melt(data, id_vars=['DStage'], value_vars=['delta_days_simphoto', 'delta_days_sim'])
        data = data.assign(fun=funname)
        dfall = pd.concat([dfall, data])
dfall.to_excel('D:/workspace/rice_crop_model/data/clusters/sites_clusters/general_parameters/error_3.xlsx',index=False)
```
