```python
import pandas as pd
import numpy as np
```

```python
thermal_df=pd.read_excel('D:/workspace/rice_crop_model/data/dfall.xlsx',)
```

```python
thermal_df.index = thermal_df['Date']
```

```python
thermal_df
```

```python
thermal_median=thermal_df['Thermal_cum'].groupby(thermal_df['DStage']).median()
```

```python
thermal_median.plot(kind='bar')
```

```python
thermal_median
```

```python
thermal_sum = thermal_df['Thermal_cum']
```

```python
bins = [0, 137.792979, 559.765610, 857.311757, 1018.537240, 1493.815289, 2000]
labesl = ['reviving data','tillering data','jointing data','booting data','heading data','maturity data']
```

```python
thermal_df['new_dvs']=pd.cut(thermal_sum,bins=bins,labels=labesl)
```

```python
thermal_df['old_dvs'] = thermal_df['DStage'].ffill()
```

```python
thermal_df['standard'] = thermal_df['old_dvs'].map(thermal_median.to_dict()) 
```

```python
thermal_df
```

```python
thermal_df['DateRev']=thermal_df['DStage']
```

```python
for i in range(thermal_df.shape[0]):
    if thermal_df['DateRev'][i]=='reviving data':
        thermal_df['DateRev'][i]=thermal_df['Date'][i]
```

```python
thermal_df['DateRev']==null
```

```python
origin = thermal_df['standard'].groupby([thermal_df['station ID'],thermal_df.index.year,thermal_df['DateRev'],thermal_df['old_dvs']]).count()
```

```python
origin
```

```python
# newdf['new_dvs'] = newdf.new_dvs.astype('str')
```

```python
df_select = thermal_df.copy()
```

```python
df_select['new_dvs'] = df_select.new_dvs.astype('str')
```

```python
df_select
```

```python
new_map = {}
for i,itm in enumerate(df_select.new_dvs.unique()):
    new_map[itm] = i
```

```python
new_map
```

```python
df_select['new_standard'] = df_select['new_dvs'].map(new_map)
```

```python
df_select
```

```python
simulate = df_select['new_standard'].groupby([df_select['station ID'],df_select.index.year,df_select['new_dvs']]).count()
```

```python
simulate
```

```python
origin
```

```python
results = simulate.unstack() - origin.unstack()
```

```python
results = results.reset_index()
```

```python
results = results.drop([ 'nan', 'maturity data' ],axis=1)
```

```python
results
```

```python
results.columns[3:]
```

```python
order = ['station ID','Date','reviving data', 'tillering data', 'jointing data', 'booting data', 'heading data']
results = results[order]
```
